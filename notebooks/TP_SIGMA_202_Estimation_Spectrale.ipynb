## Processus AR

#### Question 1.1

```python
import numpy as np
from scipy.stats import norm
from scipy.linalg import norm as norm2
from scipy.signal import hamming, boxcar, lfilter, welch, periodogram
from scipy.fftpack import fft
import matplotlib.pyplot as plt

%matplotlib inline

def AR(N, phi, sigma):  
    m = 400
    Z = sigma * norm.rvs(size = m + N)
    X = lfilter([1], phi, Z)
    return X[400:]


def autocovariance(Xi, N, k, Xs):
    """méthode prise sur internet"""
    autoCov = 0
    for i in np.arange(N - k):
        autoCov += ((Xi[i + k]) - Xs) * (Xi[i] - Xs)
    return 1. / (N - 1) * autoCov


def covariance(X):
	n = len(X)
	cov = np.zeros(n)
	esp_X = np.mean(X)
    
	for i in np.arange(n):
		cov[i] = autocovariance(X, n, i, esp_X)
        
	return cov


def DSP(X, M):
    cov = covariance(X)
    return np.real(fft(cov, M))


def estimateur_classique(phi, M):
    tf = fft(phi, M)
    out = [1. / (np.abs(tf[k]) ** 2) for k in np.arange(M)]
    
    return out
    

def periodogramme(X, M):
    #n = len(X)
    #Xc = X - np.mean(X) * np.ones(n)
    #out = 1. / n * np.abs(fft(Xc, M)) ** 2
    
    out = periodogram(X, nfft = M)[1]
    
    return out
    
    
def welsh(X, M, window = hamming(64)):
    #n = len(X)
    #Xc = X - np.mean(X) * np.ones(n) 
    #m = len(window)
    #U = 1. / m * norm2(window, 2)
    #welch = 0     
    #
    #for k in np.arange(K):
    #    Y = [window[i] * Xc[i + k * D] for i in np.arange(min(m, len(Xc[k * D:])))]
    #    welch += np.abs(ft.fft(Y, M)) ** 2
    #    
    #out *= 1. / (K * n * U)
    
    out = welch(X, window = window, nfft = M)[1]
    
    return out
```

#### Question 1.2

```python
M = 512.
mu = np.arange(M) / M
sigma = 0.2
phi = [1, - 0.5]
n = 256.
m = M / 2
X = AR(n, phi, sigma)

mu = mu[:m]

dsp = DSP(X, M)[:m]
per = periodogramme(X, M)[:m]
bar = welsh(X, M, boxcar(64))[:m]
wel = welsh(X, M, hamming(64))[:m]

plt.plot(mu, 20*np.log10(np.absolute(per)), 'r',
         mu, 20*np.log10(np.absolute(dsp)))
plt.title('Periodogramme')
plt.xlabel('Frequences reduites')
plt.ylabel('DSP en dB')
plt.grid(True)
axes = plt.gca()
axes.set_ylim((-50,0))
plt.show()

biais = np.real(np.mean([np.mean(per) - dsp[i] for i in np.arange(m)]))
variance = np.mean([per[i] ** 2 for i in np.arange(m)]) - np.mean(per) ** 2

print('Biais du périodogramme = ' + str(biais))
print('Variance du périodogramme = ' + str(variance) + '\n')

plt.plot(mu, 20*np.log10(np.absolute(bar)), 'r', 
         mu, 20*np.log10(np.absolute(dsp)))
plt.title('Methode de Bartlett')
plt.xlabel('Frequences reduites')
plt.ylabel('DSP en dB')
plt.grid(True)
axes = plt.gca()
axes.set_ylim((-50,0))
plt.show()

biais = np.real(np.mean([np.mean(bar) - dsp[i] for i in np.arange(m)]))
variance = np.mean([bar[i] ** 2 for i in np.arange(m)]) - np.mean(bar) ** 2

print('Biais de la méthode de Bartlett = ' + str(biais))
print('Variance de la méthode de Bartlett = ' + str(variance) + '\n')
         
plt.plot(mu, 20*np.log10(np.absolute(wel)), 'r',
         mu, 20*np.log10(np.absolute(dsp)))
plt.title('Methode de Welch')
plt.xlabel('Frequences reduites')
plt.ylabel('DSP en dB')
plt.grid(True)
axes = plt.gca()
axes.set_ylim((-50,0))
plt.show()

biais = np.real(np.mean([np.mean(wel) - dsp[i] for i in np.arange(m)]))
variance = np.mean([wel[i] ** 2 for i in np.arange(m)]) - np.mean(wel) ** 2

print('Biais de la methode de Welch = ' + str(biais))
print('Variance de la methode de Welch = ' + str(variance) + '\n')
```

La méthode de Bartlett réduit fortement la variance de l'estimation par rapport au périodogramme. Le biais est réduit.
La méthode de Welch réduit encore plus la variance mais on perd en résolution. Certains pics que l'on voyait avant sont maintenant aplatis.

Il y a donc un prix biais et variance par rapport à la résolution du signal.

## Somme de sinus bruités

#### Question 2.1

```python
def somme_sinus(N, K, freq, ampl, phase, sigma):
    out = np.zeros(N)
    t = np.arange(N, dtype = 'double') / (N)
    for i in np.arange(K):
        out += ampl[i] * np.sin(2 * np.pi * t * freq[i] + phase[i])
        
    out += sigma * norm.rvs(size = N)
    return out
    
```

#### Question 2.4 - 6

```python
from scipy.linalg import inv, toeplitz

def capon(X, M):
    cov = covariance(X)
    R = inv(toeplitz(cov))  
    r = R[1,:]
    
    ft = fft(r, M)
      
    out = [1. / np.abs(ft[i]) ** 2 for i in np.arange(M / 2)]
    
    return out
```

## Processus ARMA

#### Question 3.1

```python
def ARMA(N, phi, theta, sigma):  
    m = 400
    Z = sigma * norm.rvs(size = m + N)
    X = lfilter(theta, phi, Z)
    return X[400:]
```

#### Question 3.2

#### Question 3.3 - 4

```python
def durbin(X, M, p, q, sigma, L):
    #On se décale sur les autocorrélations pour ne pas prendre des coéfficients pollués par la partie MA
    phi_est = lpc(X[q:], p)[0]
    phi_est[1:] = - phi_est[1:]
    temp = fft(phi_est, M)
    dsp_AR = [np.abs(temp[i]) ** 2 for i in np.arange(M/2)]
    
    #Estimation de la partie MA par un AR d'ordre L pour respecter L >> q (on prend q dans [1,5])
    phi_est2  = lpc(X, L)[0]
    #phi_est2[1:] = -phi_est2[1:] 
    sigma2 = np.mean([(X[i] - np.mean(X)) ** 2 for i in np.arange(len(X))])
    temp = fft(phi_est2, M)
    dsp_MA = [sigma2 / (np.abs(temp[i]) ** 2) for i in np.arange(M/2)]
    
    out = [dsp_MA[i] / dsp_AR[i] for i in np.arange(int(M/2), dtype = 'int')]
    
    return out 
```

```python
L = 15

dur = durbin(X, M, 1, 1, sigma, L)

plt.plot(mu, 20*np.log10(np.absolute(dur)), 'r',
         mu, 20*np.log10(np.absolute(dsp)))
plt.title('Methode de Durbin')
plt.xlabel('Frequences reduites')
plt.ylabel('DSP en dB')
plt.grid(True)
axes = plt.gca()
axes.set_ylim((-50,100))
plt.show()

biais = np.real(np.mean([np.mean(dur) - dsp[i] for i in np.arange(m)]))
variance = np.mean([dur[i] ** 2 for i in np.arange(m, dtype = 'int')]) - np.mean(dur) ** 2

print('Biais de la methode de Durbin = ' + str(biais))
print('Variance de la methode de Durbin = ' + str(variance) + '\n')
```

La méthode de Durbin offre un estimateur bien plus adapté pour estimer la DSP d'un processus ARMA par rapport aux estimateurs d'un processus AR. La partie MA joue un rôle déterminant dans la dégradation de la DSP du ARMA par rapport à la DSP du AR simple.
